```python
X=iris['data'][:,3:]
y=(iris['target']==2).astype(np.int_)
y
```

```python
from sklearn.linear_model import LogisticRegression
log_res=LogisticRegression()
log_res.fit(X,y)
```

```python
X_new=np.linspace(0,3,1000).reshape(-1,1)
y_proba=log_res.predict_proba(X_new)
```

```python
y_proba
```

```python
plt.figure(figsize=(12,5))
decision_boundary=X_new[y_proba[:,1]>=0.5][0]
plt.plot([decision_boundary,decision_boundary],[-1,2],'k:',linewidth=2)
plt.plot(X_new,y_proba[:,1],'g--',label='Iris-Virginica')
plt.plot(X_new,y_proba[:,0],'b-',label='not Iris-Virginica')
plt.arrow(decision_boundary[0],0.08,-0.3,0,head_width=0.05,head_length=0.1,fc='b',ec='b')
plt.arrow(decision_boundary[0],0.88,0.3,0,head_width=0.05,head_length=0.1,fc='g',ec='g')
plt.text(decision_boundary[0]-0.4,0.15,'Decision_Boundary',fontsize=16,color='k')
plt.xlabel('Peta width(cm)',fontsize=16)
plt.ylabel('y_proba',fontsize=16)
plt.axis([0,3,-0.02,1.02])
plt.legend(loc='center left',fontsize=16)
```

```python
softmax_reg.predict_proba([[5,2]])
```

```python
x0,x1=np.meshgrid(np.linspace(0,8,500).reshape(-1,1),np.linspace(0,3.5,200).reshape(-1,1))
X_new=np.c_[x0.ravel(),x1.ravel()]
y_proba=softmax_reg.predict_proba(X_new)
y_predict=softmax_reg.predict(X_new)
zz1=y_proba[:,0].reshape(x0.shape)
zz=y_predict.reshape(x0.shape)
```

```python
from sklearn.datasets import make_blobs
blob_centers=np.array([
[0.2,2.3],
    [-1.5,2.3],
    [-2.8,1.8],
    [-2.8,2.8],
    [-2.8,2.3]
])
blob_std=np.array([0.4,0.3,0.1,0.1,0.1])
```

```python
X,y=make_blobs(n_samples=2000,centers=blob_centers,cluster_std=blob_std,random_state=7)
```

```python
def plot_cluster(X,y=None):
    plt.scatter(X[:,0],X[:,1],c=y,s=1)
    plt.xlabel('$x_1$',fontsize=14)
    plt.ylabel('$x_2$',fontsize=14,rotation=0)
plt.figure(figsize=(8,4))
plot_cluster(X)
plt.show()
```
